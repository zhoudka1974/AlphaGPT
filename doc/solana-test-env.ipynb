执行基本操作
1, 转账（发送 SOL）
向另一个地址发送 0.1 SOL：
    solana transfer <RECIPIENT_ADDRESS> 0.1
    例如：
        solana transfer 8Ww3Y7K5wuZzF4XqoV9HJk7GpUv9T7hQaRbNnMmZzXyA 0.1
2, 查看交易历史
    solana transaction-history <YOUR_ADDRESS>
3, 查看账户信息
    solana account <ADDRESS>
4, 部署智能合约（Program）——进阶
如果你有 .so 格式的程序（BPF ELF 文件），可以部署：
    solana program deploy your_program.so
    部署后会返回 Program ID（即合约地址）。

其他常用命令
    solana cluster-version	查看当前集群版本
    solana gossip	查看网络中的验证者节点
    solana validators	列出活跃验证者
    solana block-height	查看当前区块高度

1. Triton One（你提到的 rpcpool）
    URL: https://devnet.rpcpool.com
    特点：无需 API Key，免费使用，由 Triton One 运营（你日志中提到的服务）
    SOLANA_DEVNET_RPC_URL = "https://devnet.rpcpool.com"
    ⚠️ 注意：虽然公开可用，但建议注册获取专属 endpoint（带 token）以获得更高配额和稳定性。
2. Helius（推荐，高性能）
    官网: https://helius.dev
    免费 tier 支持 Devnet
    注册后获取 endpoint 格式：
    https://devnet.helius-rpc.com/?api-key=YOUR_API_KEY
    优点：低延迟、支持 Webhook、增强 API（如 token metadata）
3. QuickNode
    官网: https://www.quicknode.com/chains/solana
    提供 Devnet 节点
    免费账户可创建 endpoint，格式如：
    https://solana-devnet.quiknode.pro/your-api-key/
4. Alchemy
    官网: https://www.alchemy.com/solana
    支持 Mainnet Beta 和 Devnet
    免费 tier 可用，endpoint 类似：
    https://solana-devnet.g.alchemy.com/v2/YOUR_API_KEY
