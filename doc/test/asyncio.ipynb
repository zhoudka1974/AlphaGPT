```python
import asyncio

async def say_hello():
    print("Hello")
    await asyncio.sleep(1)  # 模拟异步 I/O 操作
    print("World")

# Python 3.7+ 推荐方式
# asyncio.run(say_hello())
await say_hello()
```

```python
import asyncio

async def greet(name, delay):
    print(f"开始问候 {name}")
    await asyncio.sleep(delay)
    print(f"Hello, {name}!")

async def main():
    # 并发运行三个任务
    await asyncio.gather(
        greet("Alice", 2),
        greet("Bob", 1),
        greet("Charlie", 1.5)
    )

# asyncio.run(main())
await main()
```
